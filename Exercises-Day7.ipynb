1. String - String contains only characters that are written inside "" or '' or ''' ''' , """ """ for multiline strings
2. List - List is a collection of different data types items , it is modifiable and we can add , remove and update the items in a list using the indexing, it also has slicing where we can slice out a part of the list using indexing [:] or [::]
3. Tuple - Tuple is collection which is not modifiable and it can store only a single data type in a single set , we cannot add, update or remove items from a tuple , we can use indexing and slicing to access the items in a tuple but cannot change the values
4. A set is a collection which stores only different items similar to mathematical set, we it is defined by {"item','item'} and set() method, the set stores items in an unordered manner , it contains an item only a single time no duplicate items, we can add , remove items in a set, but we cannot use indexing to access the items of the set, we use mathematical set operations on sets like union , intersection, subset, superset, disjoint, summetric difference between sets

```python
sentence = 'I am a teacher and I love to inspire and teach people'
sen_set = set(sentence.split())
```
