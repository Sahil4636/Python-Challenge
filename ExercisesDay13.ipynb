# Filter only negative and zero in the list using list comprehension

```python
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]
filetered_list = [i for i in numbers if i<= 0]
print(filetered_list)
```

```python
list_of_lists =[[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened_list = [number for row in list_of_lists for number in row  ]
print(flattened_list)
```

```python
result = [tuple([i] + [i**j for j in range(6)]) for i in range(11)]
print(result)
```

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]

result2 = [[country.upper(), country[:3].upper(), city.upper()] for sublist in countries for country, city in sublist]
print(result2)
```

```python
result3 = [{'country':country.upper(), 'city': city.upper()} for sublist in countries for country, city in sublist ]
print(result3)
```
