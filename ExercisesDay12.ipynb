```python
import random
import string

def random_user_id():
    characters = string.ascii_letters + string.digits
    user_id = ''

    for i in range(6):
        user_id += random.choice(characters)

    return user_id
print(random_user_id())
```

```python
def user_id_gen_by_user():
    a = int(input("Enter Number of Characters:"))
    b = int(input("Enter numbers of IDs:"))
    characters = string.ascii_letters + string.digits
    

    for i in range(b):
        user_id = ""

        for j in range(a):
            user_id += random.choice(characters)

        print(user_id)

user_id_gen_by_user()

    
```
