```python
import re
```

```python

def find_most_common_words(url, a):
    response = requests.get(url)
    text = response.text
    words = re.findall(r'\b[a-zA-Z]+\b', text.lower())

    freq = {}

    for word in words:
        
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    sorted_words = sorted(freq.items(), key=lambda x: x[1], reverse = True)
    formated = [(count, word) for word, count in sorted_words]

    return formated[:a]
```

```python
find_most_common_words('https://www.google.com/robots.txt', 10)
```

```python
import requests
import statistics
from collections import Counter


cats_api = 'https://api.thecatapi.com/v1/breeds'

response = requests.get(cats_api)
cats = response.json()

weights = []
lifespans = []
countries = []

for cat in cats:
    weight = cat['weight']['metric']

    try:
        low, high = map(float, weight.split(' - '))
        avg_weight = (low + high) /2
        weights.append(avg_weight)
    except:
        pass

    life = cat['life_span']

    try:
        low, high = map(int, life.split(' - '))
        avg_life = (low + high)/2
        lifespans.append(avg_life)
    except:
        pass

    countries.append(cat['origin'])

print("WEIGHT STATISTICS")
print("Min :", min(weights))
print("Max :", max(weights))
print("Mean :", statistics.mean(weights))
print("Median :", statistics.median(weights))
print("Standard Deviation :", statistics.stdev(weights))

print()

print("LIFESPAN STATISTICS")
print("Min :", min(lifespans))
print("Max :", max(lifespans))
print("Mean :", statistics.mean(lifespans))
print("Median :", statistics.median(lifespans))
print("Standard Deviation :", statistics.stdev(lifespans))

print()

print('Country Frequency')

country_count = Counter(countries)

for country, count in country_count.items():
    print(f'{country:20} {count}')
```
