```python
lambda num: num **2 # this lambda function squares a number
```

1. Lambda function can have multiple parameters
2. it should be written in a single line of code
3. it has syntax different from regular functions - they don't need written statement , they automatically written the operator part 
4. they don't have a name , they are also called annonymous functions

```python
# Normal way
def square(num):
    return num**2
def cube(num):
    return num**3

def transform_list(nums_list, transform_item):
    transform_0 = transform_item(nums_list[0])
    transform_1 = transform_item(nums_list[1])
    return [transform_0, transform_1]

my_list = [2, 3]

transform_list(my_list , square)
transform_list(my_list, cube)
```

# In the above code the the transform_list function is called a Higher-Order function

```python
def transform_list(nums_list, transform_item):
    transform_0 = transform_item(nums_list[0])
    transform_1 = transform_item(nums_list[1])
    return [transform_0, transform_1]

transform_list(my_list, lambda num: num**2)
```

# The lambda function reduces the need for writting the function separately

# map()

```python
nums_list = [2,3,4,5,6]
map(lambda num:num**3, nums_list)
```

```python
list(map(lambda num:num**2, nums_list))
```

# filter()

```python
list(filter(lambda num: num%2 == 0, nums_list))
```

```python
words = ['hi', 'hello', 'sun', 'amazing', 'cat']
long_words = filter(lambda i: len(i)>4, words)
print(list(long_words))
```

```python
students = [('Alice', 88), ('Bob', 72), ('Charlie', 95)]
sorted_students = sorted(students, key = lambda x: x[1], reverse = True)
print(sorted_students)
```

```python
people = [
    {"name" : 'Sahil', 'age':21, 'occupation':'Engineer'},
    {"name" : 'Pranay', 'age':30, 'occupation':'Engineer'},
    {"name" : 'Pranav', 'age':22, 'occupation':'Engineer'},
    {"name" : 'Kunal', 'age':22, 'occupation':'Engineer'}
]
```
