To scrape data from websites, basic understanding of HTML tags and CSS selectors is needed. We target content from a website using HTML tags, classes or/and ids. Let us import the requests and BeautifulSoup module

```python
import requests 
from bs4 import BeautifulSoup
```

```python
url = 'https://archive.ics.uci.edu/dataset/53/iris'

response = requests.get(url)

status = response.status_code
print(status) # 200 means the fetching was successful
```

Using beautifulSoup to parse content fron the page

```python
url = "https://www.w3schools.com/html/html_tables.asp"


response = requests.get(url)
content = response.content # we get all the content from the website
soup = BeautifulSoup(content, 'html.parser') # beautiful soup will give a chance to parse
print(soup.title)  # Prints title of the webpage
print(soup.title.get_text()) # printts only text of the title
 # print(soup.body)
print(response.status_code) 

table = soup.find('table', {'id':'customers'})
for row in table.find_all('tr'):
    cols = row.find_all(['th', 'td'])
    print([col.get_text(strip=True) for col in cols])
```
