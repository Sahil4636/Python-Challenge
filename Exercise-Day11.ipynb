```python
def add_two_numbers(a, b):
    return a+b
print(add_two_numbers(10,12))
```

```python
def area_of_circle(r):
    pi = 3.14
    area = pi*r**2
    return area
print(area_of_circle(30))
```

```python
def add_all_nums(*nums):
    total = 0
    for num in nums:
        if type(num) == int or type(num) == float:
            total += num
        else:
            return "Enter Valid Number"  
    return total
print(add_all_nums(2,3,4,5,6,7,8,9))
```

```python
def celsius_to_fehreheit(c):
    F = (c*9/5) + 32
    return F
print(celsius_to_fehreheit(30))
```

```python
def check_season(month):
    if month == "September" or month == 'October' or month == 'November':
        return "The season is Autumn."
    elif month == "December" or month == "January" or month == 'February':
        return "The season is Spring."
    elif month == 'June'or month == 'July'or month == "August":
        return "The Season is Summer"
    else:
        return "please enter valid month"
print(check_season("September"))
```

```python
def calculate_slope(x1, x2, y1, y2):
    slope = (y2-y1)/(x2-x1)
    return slope
print(calculate_slope(2,4,3,7))
```

```python
def solve_quadratic_eqn(a, b, c):
    discriminant = b**2 - 4*a*c

    if discriminant < 0:
        return "No real solutions"

    x1 = (-b + discriminant**0.5)/(2*a)
    x2 = (-b - discriminant**0.5)/(2*a)

    return (x1, x2)

print(solve_quadratic_eqn(1,-3,2))
```

```python
def print_list(list):
    for i in range(len(list)):
        print(list[i])
        
print(print_list([2,4,5,"sahil", "dhanvij"]))
```

```python
def reverse_list(list):
    return list[::-1]

print(reverse_list([2,4,5,"sahil", "dhanvij"]))
```

```python
def capitalize_list_items(list):
    for i in range(len(list)):
        list[i] = list[i].capitalize()
    return list
print(capitalize_list_items(['sahil', 'pranay', 'pranav', 'atharv', 'aman']))
```

```python
def add_item(list, item):
    list.append(item)
    return list
print(add_item(['Potato', 'Tomato', 'Mango', 'Milk'], 'Meat'))
```

```python
def remove_item(list, item):
    list.remove(item)
    return list
print(remove_item(['Potato', 'Tomato', 'Mango', 'Milk'], 'Milk'))
```

```python
def sum_of_numbers(num):
    total = 0
    for i in range(num +1):
        total += i
    return total
print(sum_of_numbers(100))
```

```python
def sum_of_odds(num):
    total = 0
    for i in range(num+1):
        if i%2 !=0 :
            total += i
    return total
print(sum_of_odds(100))
```

```python
def sum_of_even(num):
    total = 0
    for i in range(num+1):
        if i%2==0:
            total += i
    return total
print(sum_of_even(100))
```
