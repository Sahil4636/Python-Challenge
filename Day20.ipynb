# Reading From URL
By now you are familiar with how to read or write on a file located on you local machine. Sometimes, we would like to read from a website using url or from an API. API stands for Application Program Interface. It is a means to exchange structured data between servers primarily as json data. To open a network connection, we need a package called requests - it allows to open a network connection and to implement CRUD(create, read, update and delete) operations. In this section, we will cover only reading ore getting part of a CRUD.

```python
pip install requests
```

We will see get, status_code, headers, text and json methods in requests module:

1. get(): to open a network and fetch data from url - it returns a response object
2. status_code: After we fetched data, we can check the status of the operation (success, error, etc)
3. headers: To check the header types
4. text: to extract the text from the fetched response object
5. json: to extract json data Let's read a txt file from this website,https://www.w3.org/TR/PNG/iso_8859-1.txt

```python
import requests
url = 'https://www.w3.org/'

response = requests.get(url)
print(response)
print(response.status_code)
print(response.headers['Content-Type'])
print(response.text[:500])
```

Let us read from an API. API stands for Application Program Interface. It is a means to exchange structure data between servers primarily a json data. An example of an API:https://jsonplaceholder.typicode.com/users/1"
. Let us read this API using requests module.

```python
import requests

url = "https://jsonplaceholder.typicode.com/users/1"

response = requests.get(url)

print(response.status_code)

data = response.json()

print(data["name"])
print(data["email"])
```

We use json() method from response object, if the we are fetching JSON data. For txt, html, xml and other file formats we can use text.
