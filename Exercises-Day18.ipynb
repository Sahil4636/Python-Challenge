```python
paragraph = 'I love teaching. If you do not love teaching what else can you love. I love Python if you do not love something which can give you all the capabilities to develop an application what else can you love.'
```

```python
from collections import Counter
import re

words = re.findall(r'\b\w+\b', paragraph, re.I)

counts = Counter(words)
print(counts.most_common(1))

```

Revise

```python
para = paragraph.lower()
para = para.replace('.', '')
para = para.replace(',', '')
words = para.split()

def most_frequent(pa):
    freq = {}
    for i in words:
        if i not in freq:
            freq[i]= 1
        else:
            freq[i] += 1
    max_word = ""
    max_count = 0
    for word, count in freq.items():
        if count > max_count:
            max_count = count
            max_word = word
    return (max_word, max_count)

most_frequent(para)
```

```python
points = ['-12', '-4', '-3', '-1', '0', '4', '8']
points = [int(i) for i in points ]
print(points)
furt = len(points) - 1
distance = points[furt] - points[0]
print(f'distance = {points[furt]} - ({points[0]}) = {distance}')
```

```python
text = """The position of some particles on the horizontal x-axis are -12, -4, -3 and -1
in the negative direction, 0 at origin, 4 and 8 in the positive direction."""

points = re.findall(r'-?\d+', text)
print(points)
numbers = list(map(int, points))
distance = max(numbers) - min(numbers)
print(distance)
```

```python
def is_valid_var():
    variable = input("Enter variable name: ")
    if re.match(r'^[a-zA-z_][a-zA-Z0-9_]*$', variable):
        print('Valid')
    else:
        print("Invalid")

is_valid_var()
```
